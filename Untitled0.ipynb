```python
# Step 1: Import libraries

import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset

df = pd.read_csv("/content/sales_data_sample.csv", encoding='latin-1')  # Ensure file exists at this path

# Step 3: Display the first few rows

df.head()
```

```python
# Step 3: Check for missing values
df.isnull().sum()
```

```python
# Step 4: Simple analysis – Total export value by state
# Remove rows with missing values in the 'STATE' column before sorting
df_cleaned = df.dropna(subset=['STATE']).copy() # Add .copy() to avoid SettingWithCopyWarning

# Convert the 'STATE' column to string type
df_cleaned['STATE'] = df_cleaned['STATE'].astype(str)

df_sorted = df_cleaned.sort_values(by='SALES', ascending=False)

# Step 5: Visualize top 10 exporting states
top10 = df_sorted.head(10)

plt.figure(figsize=(10,6))
plt.bar(top10['STATE'], top10['SALES'], color='skyblue')
plt.title("Top 10 Exporting States")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
